# Visual of Disk Annuli

Just for demonstration purposes (as well as interim reports and/or poster), making a nice plot of disk annuli. Using KHH88, but give any tables of $r_a$, $r_b$, and $T$, and this will work.

```python
import numpy as np
import matplotlib
from matplotlib.patches import Circle, Wedge, Polygon
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt
```

```python
KHH_RAD = np.array([1.5, 1.73, 1.95, 2.14, 2.26, 2.37,\
                   2.55, 2.77, 2.98, 3.20, 3.85, 4.60,\
                   4.86, 5.19, 5.70, 6.28, 6.96, 7.77,\
                   8.73, 9.91, 11.37, 13.22, 15.64,\
                   18.89, 23.46, 28.16, 32.35, 37.74,\
                   44.86, 54.64, 68.77, 90.64, 128.19,\
                   204.84, 429.26])
KHH_TEMP = np.array([6590, 6370, 6150, 5800, 5650,\
                    5500, 5370, 5100, 4900, 4700,\
                    4500, 3750, 3600, 3500, 3300,\
                    3100, 2900, 2700, 2500, 2300,\
                    2100, 1900, 1700, 1500, 1300,\
                    1100, 1000, 900, 800, 700, 600,\
                    500, 400, 300, 200])
```

```python
newlis = np.zeros(len(KHH_RAD)+1)
newlis[0] = 1.
for i in range(len(KHH_RAD)):
    newlis[i+1] = KHH_RAD[i]
```

```python
KHH_RAD_IN = newlis[:-1]
```

```python
plt.rcParams.update({'font.size': 16})

fig, ax = plt.subplots()
patches = []


for i in range(len(KHH_RAD)):
    patches.append(Wedge((0, 0), KHH_RAD[i], 0, 360., width=KHH_RAD[i] - KHH_RAD_IN[i]))

fig.set_size_inches(12, 9)
colors = KHH_TEMP
p = PatchCollection(patches, alpha=1)
p.set_array(np.array(colors))
ax.add_collection(p)
fig.colorbar(p, ax=ax, label='Temperature (K)')
plt.xlim(-500,500)
plt.ylim(-500,500)

plt.xlabel(r'$x$-displacement from central star ($R_*$)')
plt.ylabel(r'$y$-displacement from central star ($R_*$)')


plt.title('Accretion Disk with Annuli at Different Temperatures')


plt.show()
```
