## Making an instance of the class FU Ori and testing

```python
M_STAR = 0.68*M_SUN
R_STAR = 0.41*R_SUN
R_OUTER = 200 * R_STAR
INC = 15.
M_DOT = (10**-1.5) * 1e-6 * M_SUN / 3.154e7
DIST = 1.27e3*3.08567758128e18
DIREC = 'C:/Users/tonyc/OneDrive/Documents/2019 Summer/Stellar Atmospheres/'
```

```python
class FUOriEnvironment:
    
    # Initializer / Instance Attributes
    def __init__(self, r_star, r_outer, m_star, m_dot, inc, dist, direc):
        self.r_star = r_star
        self.r_outer= r_outer
        self.m_star = m_star
        self.m_dot = m_dot
        self.inc = inc
        self.dist = dist
        self.direc = direc
        
        # Storing classes
        self.fuori = []
        self.rot_star = 0
        
        print('Created environment.')
        
    # instance methods
    
    # Creating FU Ori object as accretion disk
    def createFUOri(self, grav):
        print('Creating FU Ori object...')
        fuori = FUOri(self.r_star, self.r_star, self.r_outer, self.m_star, self.m_dot, np.radians(self.inc))
        print('Creating disk annuli...')
        fuori.setAnnuliValues(10000, 2000, 200, 2*R_STAR)
        print('Reading stellar atmosphere directory...')
        fuori.createAnnuliFromDirec(grav, self.direc, model='stellar atmospheres')
        print('Preparing annuli for broadening and summing...')
        fuori.prepareAnnuli(4e3,20e3, binning=1, broaden=True, interp_type='linear')
        print('Creating model spectrum...')
        fuori.createModelSpectrum('stellar atmospheres', broaden=True)
        print('FU Ori object created.')
        self.fuori.append(fuori)
    
    # Creating rotating star
    def createRotStar(self, temp):
        self.rot_star = Star(temp, 4.0, self.m_star, self.r_star, self.direc, np.radians(self.inc)) 
        self.rot_star.setSpectrumFromDirec('stellar atmospheres', 4400, 1.5)
        self.rot_star.prepareSpectrum(4e3, 10e3, 1, 'linear')
        self.rot_star.broadenStellarSpectrum()
    
    # For viewing model spectrum
    def viewModelSpectrum(self, wave_lower, wave_upper, obj_type, obj_num, broaden, redden):
        if obj_type == 'fu ori':
            return self.fuori[obj_num].viewModelSpectrum(wave_lower, wave_upper, broaden, redden)
        elif obj_type == 'rotating star':
            return self.rot_star.getLuminosity(wave_lower, wave_upper, broaden, redden)
    
    # For viewing model SED
    def viewModelSED(self, wave_lower, wave_upper, broaden):
        waves, lums = self.fuori.viewModelSpectrum(wave_lower, wave_upper, broaden, redden)
        return np.log10(waves*1e-4), np.log10(np.cos(np.radians(self.inc))*waves*1e-8*lums/(2*np.pi*(self.dist**2)))

    # Useful modules
    def viewFracContributions(self, bands, horiz_axis):
        pass
    
    def viewGravComparison(self, obj_nums, wave_lower, wave_upper, broaden, redden):
        obj_n1, obj_n2 = obj_nums
        wav, lum1 = self.viewModelSpectrum(wave_lower, wave_upper, 'fu ori', obj_n1, broaden, redden)
        wav, lum2 = self.viewModelSpectrum(wave_lower, wave_upper, 'fu ori', obj_n2, broaden, redden)
        lum_diff = np.abs(lum2 - lum1)
        return wav, lum_diff    
```

### Reading in infrared SED from Hillenbrand et al. 2018

```python
PATH = 'C:/Users/tonyc/OneDrive/Documents/2019 Summer/Papers/Gaia 17bpi/'

DIST = 1.27e3*3.08567758128e18
```

```python
def getDataArray(data_file):
    txt = open(data_file)
    fulltxt = txt.readlines()
    # Cleaning lines
    newtxt = []
    for i in range(len(fulltxt)):
        line = fulltxt[i][:-1]
        line = line.split()
        if line[0] == '#':
            line = [line[1], line[2]]
        newtxt.append(line)
    # Casting as floats
    newtxt = np.array(newtxt).astype(np.float64)
    return newtxt
```

```python
newtxt = getDataArray(PATH + 'Gaia17bpi.merged.txt')

wavesDATA_INF = newtxt[:,0] 
fluxesDATA_INF = newtxt[:,1]
```

```python
REGIONS = np.array([(1e3, 8.5e3), (13e3, 15e3), (17.5e3, 19.8e3), (24.2e3, 26.5e3)])
```

```python
def cleanPoints(waves, lums, regions):
    waves_final, lums_final = [], []
    exclude = []
    for i in range(len(regions)):
        start, end = regions[i]
        ind_lower = np.searchsorted(waves, int(start))
        ind_upper = np.searchsorted(waves, end)
        bad_points = np.arange(ind_lower, ind_upper+1, 1)
        for i in range(len(bad_points)):
            exclude.append(bad_points[i])

    waves_final = np.delete(waves, exclude)
    lums_final = np.delete(lums, exclude)
    return waves_final, lums_final
```

```python
waves_DAT, fluxes_DAT = cleanPoints(wavesDATA_INF, fluxesDATA_INF, REGIONS)
```

```python
fig=plt.figure(figsize=(6, 6), dpi= 80)
plt.rcParams.update({'font.size': 14})

plt.xlabel(r'log $\lambda$ ($\mu$m)')
plt.ylabel(r'log $\lambda F_\lambda (erg\quad s^{-1} cm^{-2})$')

plt.title('Gaia 17bpi Infrared SED data points')

plt.plot(np.log10(wavesDATA_INF*1e-4), np.log10(fluxesDATA_INF * wavesDATA_INF), 'ro', label='Original points')
plt.plot(np.log10(waves_DAT*1e-4), np.log10(fluxes_DAT * waves_DAT), 'go', label='Cleaned points')
plt.legend()
plt.show()
```

## Comparing different surface gravities

```python
wav0, lum0 = gaia17bpi_env.viewGravComparison((0,1), START, END, True, True)
wav, lum = gaia17bpi_env.viewModelSpectrum(START, END, 'fu ori', 0, True, True)
wav2, lum2 = gaia17bpi_env.viewModelSpectrum(START, END, 'fu ori', 1, True, True)
```

```python
fig=plt.figure(figsize=(10, 7), dpi= 80)
gs = gridspec.GridSpec(2, 1,height_ratios=[3, 1])

ax = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])

ax.plot(wav, lum*np.cos(np.radians(INC))/(2*np.pi*DIST**2), label='Stationary disk, g=1.5')
ax.plot(wav2, lum2*np.cos(np.radians(INC))/(2*np.pi*DIST**2), label='Stationary disk, g=4.0')
ax.legend()
ax2.plot(wav0, lum0*np.cos(np.radians(INC))/(2*np.pi*DIST**2), 'r-', label='Absolute value of difference')
ax2.plot(wav0, np.mean(lum0*np.cos(np.radians(INC))/(2*np.pi*DIST**2))*np.ones(len(lum0)), 'g-', label='Mean difference')

ax2.legend()
plt.suptitle('Surface Gravity Comparison')
fig.text(0.5, 0.04, r'$\lambda$ (Angstrom)', ha='center')
fig.text(0.04, 0.5, r'$F_\lambda\; (erg \;cm^{-2} s^{-1} cm^{-1})$', va='center', rotation='vertical')
plt.show()
```
