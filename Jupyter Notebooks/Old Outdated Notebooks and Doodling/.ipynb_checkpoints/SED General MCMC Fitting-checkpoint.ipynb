The "environment"" allows you to create a blackbody FU Ori object and store it along with a rotating star to plot together if desired.

```python
class FUOriEnvironment:
    
    # Initializer / Instance Attributes
    def __init__(self, wave_lower_SED, wave_upper_SED):
        
        self.wave_lower_SED = wave_lower_SED
        self.wave_upper_SED = wave_upper_SED
        
        # Storing classes
        self.fuori = 0
        self.rot_star = 0
        
    # instance methods
    
    # Creating FU Ori object as accretion disk
    def createFUOri(self, r_star, r_outer, m_star, m_dot, inc, a_v, dist, grav, direc):
        fuori = FUOri(r_star, r_star, r_outer, m_star, m_dot, np.radians(inc), a_v)
        fuori.setAnnuliValues(30000, 2000, 100, 0.1*R_STAR)
        fuori.createAnnuliBlackbody(np.arange(self.wave_lower_SED, self.wave_upper_SED, 10))
        fuori.createModelSpectrum('blackbodies', broaden=False)
        self.fuori = fuori
        
    def createRotStar(self, temp, r_star, m_star, inc, a_v, dist, grav, direc):
        self.rot_star = Star(temp, grav, m_star, r_star, direc, np.radians(inc), a_v) 
        self.rot_star.setSpectrumBlackbody(np.arange(self.wave_lower_SED*1e-8, self.wave_upper_SED*1e-8, 10*1e-8))
    
    # For viewing model spectrum
    def viewModelSpectrum(self, wave_lower, wave_upper, redden):
            return self.fuori.viewModelSpectrum(wave_lower, wave_upper, False, redden)
    def viewModelSpectrumStar(self, wave_lower, wave_upper, redden):
            return self.rot_star.getLuminosity(wave_lower, wave_upper, 'non-interpolated', redden)
    def getModSpecPoint(self, wave, redden):
        return np.mean(self.viewModelSpectrum(wave-10, wave+10, redden)[1])
```

## MCMC Routine

Log-likelihood. This assumes Gaussian errors at the moment. 

```python
def lnlike(theta, x, y, yerr):
    m_star, r_star, m_dot= theta
    r_star = r_star*R_SUN
    
    m_dot = 10**m_dot
    m_star = 10**m_star
    m_star = m_star*M_SUN
    
    m_dot = m_dot* 1e-6 * M_SUN / 3.154e7
    try:
    
        gaia17bpi_env = FUOriEnvironment(2.5e3, 5e4)
        gaia17bpi_env.createFUOri(r_star = r_star, r_outer = R_OUTER, m_star = m_star,\
                              m_dot = m_dot, inc = INC, a_v = A_V, dist = DIST, grav = 1.5, direc = DIREC)
    
    # Accounts for any random case where the model may not be generated properly given a certain combination
    # of parameters. This has only happened once, but it's presumably when the maximum temperature is too low.
    except:
        print('Weird array error')
        gaia17bpi_env = FUOriEnvironment(2.5e3, 5e4)
        gaia17bpi_env.createFUOri(r_star = R_STAR, r_outer = R_OUTER, m_star = M_STAR,\
                              m_dot = m_dot, inc = INC, a_v = A_V, dist = DIST, grav = 1.5, direc = DIREC)
        
    
    # Converting units to properly compare model to data
    x = (10**x)*1e4
    modSpec = np.zeros(len(x))
    for i in range(len(x)):
        modSpec[i] = gaia17bpi_env.getModSpecPoint(x[i], True)

    flux = np.cos(np.radians(INC))*modSpec/(2*np.pi*DIST**2)

    model = np.log10(x*flux*1e-8)
    inv_sig = 1/yerr
    return np.sum(np.log((1/np.sqrt(2*np.pi))*inv_sig) - 0.5*((y - model)*inv_sig)**2)
```

### Setting ranges of uniform priors

The priors for mass and accertion rate are in log-space. The former in units of $M_\odot$, the latter in units of $1\times 10^{-6} M_\odot \;\text{yr}^{-1}$. The units of radius are in terms of $R_\odot$.

```python
M_STAR_LOW = -1.0
M_STAR_HIGH = 1.0
R_STAR_LOW = 0.3
R_STAR_HIGH = 7.0

M_DOT_LOW = -1.5
M_DOT_HIGH = 3.5
```

Log-prior. Uniform for radius, log-uniform for mass and accretion rate to sample different orders of magnitude equally.

```python
def lnprior(theta):
    m_star, r_star, m_dot= theta
    if M_DOT_LOW  < m_dot < M_DOT_HIGH \
    and M_STAR_LOW < m_star < M_STAR_HIGH\
    and R_STAR_LOW < r_star < R_STAR_HIGH:
        return 0.0
    return -np.inf
```

Log-posterior is the sum of log-prior and log-likelihood.

```python
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return (lp + lnlike(theta, x, y, yerr))
```

Specify the number of walkers to be at least four times the amount of parameters. Give reasonable starting guesses.

```python
ndim, nwalkers = 3, 18
m_star_start = -0.1
r_star_start = 2.0
m_dot_start = 1
start = np.array((m_star_start, r_star_start, m_dot_start))
pos = [start + 1e-2*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
x = log_lam_OPT
y_meas = log_lam_F_OPT
y_err = log_lam_F_OPT_ERR
```

Testing to see the log-posterior given the data and initial guesses

```python
lnprob((m_star_start, r_star_start, m_dot_start), x,y_meas,y_err)
```

Running the sampler

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(x, y_meas, y_err))
%time sampler.run_mcmc(pos, 1200)
```

### Plotting the chains

```python
samples = sampler.chain[:, 800:, :].reshape((-1, ndim))
fig = plt.figure(figsize=(10,6))
plt.rcParams.update({'font.size': 16})

plt.plot(samples)
plt.xlabel('Step number')
plt.ylabel('Estimated values of parameters')
plt.title('Evolution in estimated values using MCMC')
plt.xticks([])
plt.show()
```

Making the corner plot. The $\texttt{corner}$ package has a bug such that the errors reported in the text above each panel wil always correspond to the 16 and 84 percent quantiles, so I'm limited to those if I want the dotted lines to correspond to the numbers above. These are not bad quantiles to have, but if you don't care and would like different ones, just change the quantile values and ignore the text above the panels. The median value appears in red.

```python
median_calcuated = np.median(samples, axis=0)

fig = corner.corner(samples, labels=[r'$\log\; M_*$', r'$R_*$',r'$\log\;\dot{M}$'], verbose=True,
                       quantiles=(0.16, 0.5,0.84), plot_contours=True, scale_hist=True,
                       show_titles=True, title_kwargs={"fontsize": 12})

axes = np.array(fig.axes).reshape((ndim, ndim))

for i in range(ndim):
    ax = axes[i, i]
    ax.axvline(median_calcuated[i], color="r")
    
for yi in range(ndim):
    for xi in range(yi):
        ax = axes[yi, xi]
        ax.axvline(median_calcuated[xi], color="r")
        ax.axhline(median_calcuated[yi], color="r")
        ax.plot(median_calcuated[xi], median_calcuated[yi], "sr")

fig.show()
# fig.savefig("hbc722_0816_15.png", dpi=140)
```
