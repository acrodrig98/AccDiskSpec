### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy import integrate
import emcee
import corner
import time

%matplotlib inline 
```

```python
G = 6.67259e-8
SIG_SB = 5.67051e-5
M_SUN = 1.99e33
R_SUN = 6.96e10
L_SUN = 3.839e33
h_PLANCK = 6.6260755e-27
c_LIGHT = 2.99792458e10
k_BOLTZ = 1.380658e-16
```

### Temperature profile functions

```python
def tempKepDisk(r, r_inner, m_dot, m_star):
    term1 = 3*G*m_star*m_dot / (8 * np.pi * SIG_SB * (r**3))
    term2 = (1 - (r_inner/r)**(1/2))
    return (term1 * term2)**(1/4)

def tempKepDiskNeg(r, r_inner, m_dot, m_star):
    return -tempKepDisk(r, r_inner, m_dot, m_star)

def findMaxTemp(r_inner, m_dot, m_star):
    res = minimize(tempKepDiskNeg, 2*r_inner, args=(r_inner, m_dot, m_star), method='Nelder-Mead')
    return (res.x[0], -res.fun)

def tempFUOriDisk(r, r_inner, m_dot, m_star):
    # Doesn't read in as arrays for some reason,
    # doing this element-wise...
    res = np.zeros(len(r))
    for i in range(len(r)):
        if r[i] <= findMaxTemp(r_inner, m_dot, m_star)[0]:
            res[i] = findMaxTemp(r_inner, m_dot, m_star)[1]
        else:
            res[i] = tempKepDisk(r[i], r_inner, m_dot, m_star)
    return res
```

### Annuli Generators

```python
def generateDistArrays(r_inner, r_outer, r_sep):
    r_list = np.arange(r_inner, r_outer, r_sep)
    r_tup = []
    for i in range(len(r_list)-1):
        r_tup.append((r_list[i], r_list[i+1]))
    r_tup = np.array(r_tup)
    r_b = np.array([i[1] for i in r_tup])
    r_a = np.array([i[0] for i in r_tup])
    r_avg = np.mean(r_tup, axis=1)
    return (r_tup, r_avg, r_a, r_b)
```

### Luminosity functions

```python
def lumWaveAnnulus(wavelength, temp, r_b, r_a):
    term1 = 2*np.pi *(r_b**2 - r_a**2)
    term2 = 2*h_PLANCK*(c_LIGHT**2)*(wavelength**(-5))*np.pi
    term3 = (np.exp(h_PLANCK*c_LIGHT/(wavelength*k_BOLTZ*temp))-1)**(-1)
    return term1*term2*term3

def lumFreqAnnulus(freq, temp, r_b, r_a):
    term1 = 2*np.pi *(r_b**2 - r_a**2)
    term2 = 2*h_PLANCK*(c_LIGHT**(-2))*(freq**3)*np.pi
    term3 = (np.exp(h_PLANCK*freq/(k_BOLTZ*temp))-1)**(-1)
    return term1*term2*term3
```

### Class structure

```python
class FUOri:

    # Initializer / Instance Attributes
    def __init__(self, r_star, r_inner, r_outer, m_star, m_dot):
        self.r_star = r_star
        self.r_inner = r_inner
        self.r_outer= r_outer
        self.m_star = m_star
        self.m_dot = m_dot
        
        # Distance arrays
        self.distances = 0

    # instance methods
    
    # Setting the distance arrays, DO THIS FIRST!
    def makeDistArrays(self, r_sep):
        self.distances = generateDistArrays(self.r_inner, self.r_outer, r_sep)
        
    def makeListDistArrays(self, points):
        self.distances = generateListDistArrays(points)
    
    # T(r) of each annulus
    def getTemperature(self):
        r = self.distances[1]
        return tempFUOriDisk(r, self.r_inner, self.m_dot, self.m_star)

    # Luminosity contribution of each annulus
    def getLuminosityWave(self, wave):
        r_tup, r_avg, r_a, r_b = self.distances
        temps = self.getTemperature()
        return lumWaveAnnulus(wave, temps, r_b, r_a)
    
    # Model spectrum
    def getModelSpectrumWave(self, wave_lower, wave_upper, binning):
        r_tup, r_avg, r_a, r_b = self.distances
        waves = np.arange(wave_lower, wave_upper+binning, binning)
        temp = tempFUOriDisk([r_avg[0]], self.r_inner, self.m_dot, self.m_star)
        lum = lumWaveAnnulus(waves, temp, r_b[0], r_a[0])
        for i in range(1,len(r_avg)):
            temp = tempFUOriDisk([r_avg[i]], self.r_inner, self.m_dot, self.m_star)
            lum += lumWaveAnnulus(waves, temp, r_b[i], r_a[i])
        return (waves,lum)
    
    def getModelSpectrumWavePoint(self, wave):
        r_tup, r_avg, r_a, r_b = self.distances

        temp = tempFUOriDisk([r_avg[0]], self.r_inner, self.m_dot, self.m_star)
        lum = lumWaveAnnulus(wave, temp, r_b[0], r_a[0])
        for i in range(1,len(r_avg)):
            temp = tempFUOriDisk([r_avg[i]], self.r_inner, self.m_dot, self.m_star)
            lum += lumWaveAnnulus(wave, temp, r_b[i], r_a[i])
        return lum
    
    # Total luminosity
    def getTotalLuminosity(self):
        r_tup, r_avg, r_a, r_b = self.distances
        lumTot = 0
        for i in range(len(r_avg)):
            temp = tempFUOriDisk([r_avg[i]], self.r_inner, self.m_dot, self.m_star)
            lumTot += 2*np.pi*(r_b[i]**2 - r_a[i]**2)*SIG_SB*temp**4
        return lumTot[0]
    
```

### Implementing extinction

Funcational form taken from Cardelli, Clayton, and Mathis (1989). Using the ISM average of $R \approx 3.1$.

```python
def calcExtinction(x, r):
    if 0. <= x <= 1.1:
        return 0.574*(x**1.61) - (0.527*(x**1.61))/r
    elif 1.1 < x <= 3.3:
        y = x - 1.82
        return 1 + 0.17699*y - 0.50447*(y**2) - 0.02427*(y**3) + 0.72085*(y**4) \
        + 0.01979*(y**5) - 0.77530*(y**6) + 0.32999*(y**7)\
        +(1.41338*y + 2.28305*(y**2) + 1.07233*(y**3) - 5.38434*(y**4)\
        - 0.62251*(y**5) + 5.30260*(y**6) - 2.09002*(y**7) )/r
    elif 3.3 < x <= 8:
        if 5.9 <= x <= 8:
            f_a = -0.04473*(x - 5.9)**2 - 0.009779*(x - 5.9)**3
            f_b = 0.2130*(x - 5.9)**2 + 0.1207*(x - 5.9)**3
        else:
            f_a = 0
            f_b = 0
        return 1.752 - 0.316*x - 0.104/((x - 4.67)**2 + 0.341) \
                + f_a + (-3.090 + 1.825*x + 1.206/((x - 4.62)**2 + 0.263) + f_b)/r
    else:
        return 0
```

```python
def makeExtinctionCurve(waves, r):    
    waves_mod = 1/(waves*1e-4)
    return waves, np.array([calcExtinction(wave, r) for wave in waves_mod])  
```

## MCMC Routine

```python
def lnlike(theta, x, y, yerr):
    m_star, r_star, m_dot, inc = theta
    
    m_star = m_star*M_SUN
    r_star = r_star*R_SUN
    inc = np.degrees(np.arccos(inc))
    
    m_dot = 10**m_dot
    
    m_dot = m_dot* 1e-6 * M_SUN / 3.154e7
    
    gaia17bpi = FUOri(r_star, r_star, 300*R_SUN, m_star, m_dot)
    gaia17bpi.makeDistArrays(r_star)
    # Converting units
    x = (10**x)*1e-4
    modSpec = gaia17bpi.getModelSpectrumWavePoint(x)
    a_lambda = 3*makeExtinctionCurve(x*1e8, 3.1)[1]
    flux = np.cos(np.radians(inc))*modSpec/(2*np.pi*DIST**2)
    flux_corrected = flux*(10**(-a_lambda/2.5))
    model = np.log10(x*flux_corrected)
    inv_sig = 1/yerr
    return np.sum(np.log((1/np.sqrt(2*np.pi))*inv_sig) - 0.5*((y - model)*inv_sig)**2)
```

```python
M_STAR_LOW = 0.1
M_STAR_HIGH = 2.0
R_STAR_LOW = 0.1
R_STAR_HIGH = 3.0
M_DOT_LOW = -2.5
M_DOT_HIGH = 2.5
INC_LOW = 0.01
INC_HIGH = 0.99
```

```python
def lnprior(theta):
    m_star, r_star, m_dot, inc = theta
    if M_STAR_LOW < m_star < M_STAR_HIGH \
    and M_DOT_LOW  < m_dot < M_DOT_HIGH \
    and R_STAR_LOW < r_star < R_STAR_HIGH\
    and INC_LOW < inc < INC_HIGH:
        return 0.0
    return -np.inf
```

```python
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return (lp + lnlike(theta, x, y, yerr))
```

```python
ndim, nwalkers = 4, 16
m_star_start = 0.5
r_star_start = 0.6
inc_start = 0.5
start = np.array((m_star_start, r_star_start, m_dot_start, inc_start))
pos = [start + 1e-2*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
x = log_lam_OPT
y_meas = log_lam_F_OPT
y_err = log_lam_F_OPT_ERR
```

```python
lnprob((m_star_start, r_star_start, m_dot_start, inc_start), x,y_meas,y_err)
```

```python
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(x, y_meas, y_err))
%time sampler.run_mcmc(pos, 2)
```

```python
samples = sampler.chain[:, 0:, :].reshape((-1, ndim))
fig = plt.figure(figsize=(10,6))
plt.plot(samples)
plt.xlabel('Step number up to 1000 total steps')
plt.ylabel('Estimated values of parameters')
plt.title('Evolution in estimated values using MCMC')
plt.xticks([])
plt.show()
```

```python
median_calcuated = np.median(samples, axis=0)

fig = corner.corner(samples, labels=[r'$M_*$',  r'$R_*$', r'$\log\;\dot{M}$', r'$i$'], verbose=True,
                       quantiles=(0.16, 0.5,0.84), plot_contours=True, scale_hist=True,
                       show_titles=True, title_kwargs={"fontsize": 12})

axes = np.array(fig.axes).reshape((ndim, ndim))

for i in range(ndim):
    ax = axes[i, i]
    ax.axvline(median_calcuated[i], color="r")
    
for yi in range(ndim):
    for xi in range(yi):
        ax = axes[yi, xi]
        ax.axvline(median_calcuated[xi], color="r")
        ax.axhline(median_calcuated[yi], color="r")
        ax.plot(median_calcuated[xi], median_calcuated[yi], "sr")

fig.show()
# fig.savefig("gaia17bpi_0807.png", dpi=140)
```
