```python
ATM_TABLE = dict()
ATMOS_FACTOR = 100
```

```python
def getDataArrayMOD(data_file):
    txt = open(data_file)
    fulltxt = txt.readlines()
    # Cleaning lines
    newtxt = []
    for i in range(0,len(fulltxt), 2):
        line = fulltxt[i][:-1]
        line = line.split()
        if line[0] == '#':
            line = [line[1], line[2]]
        newtxt.append(line)
    # Casting as floats
    newtxt = np.array(newtxt).astype(np.float64)
    return newtxt
```

```python
def direcToArray(directory):
    for temp in range(1000, 10000+ATMOS_FACTOR, ATMOS_FACTOR):
        for grav in [1.5]:
            try:
                atmos_file = directory + 'pr.lte' + str(int(temp/ATMOS_FACTOR)) \
                        + '-' + str(grav) + '-0.0.spec'
#                 print(np.loadtxt(atmos_file)[0])
                txt = open(atmos_file)
                fulltxt = txt.readlines()
                # Cleaning lines
                newtxt = []
                for i in range(len(fulltxt)):
                    line = fulltxt[i][:-1]
                    line = line.split()
                    newtxt.append(line)
        
                # Casting as floats
                newtxt = np.array(newtxt).astype(np.float64)
                
                # Choosing which model to use
                wav, lum = newtxt[:,0], newtxt[:,1]
                
                # For DUSTY Models
                if temp < 2000:
                    lum *= 1e8
                ATM_TABLE[(temp, grav)] = np.array([wav, lum])
            except:
                print('No atmosphere found for temp = ' + str(temp) + \
                      ' grav = ' + str(grav))
```

```python
def getInterpTable(st_atm_table, waves_data):
    out_table = dict()
    for key in ATM_TABLE:
        wav, lum = ATM_TABLE[key]
#         interp_func = interpolate.interp1d(wav, lum, kind = 'linear')
#         out_table[key] = waves_data, interp_func(waves_data)
        out_table[key] =  waves_data, spectres.spectres(waves_data, wav, lum)
    return out_table
```

```python
def convertToPwrLaw(temps, r, r_star):
    return temps*(r/r_star)**(1/4)
```

```python
class FUOriEnvironment:
    
    # Initializer / Instance Attributes
    def __init__(self, wave_lower_SED, wave_upper_SED):
        
        self.wave_lower_SED = wave_lower_SED
        self.wave_upper_SED = wave_upper_SED
        
        # Storing classes
        self.fuori = []
        self.star = []
        
    # instance methods
    
    # Creating FU Ori object as accretion disk
    def createFUOri(self, r_star, r_outer, m_star, m_dot, inc, dist, grav, direc, a_v, atm_table, pl_index, mod):
        fuori = FUOri(r_star, r_star, r_outer, m_star, m_dot, np.radians(inc), a_v, atm_table, pl_index)
        fuori.setAnnuliValues(10000, 1000, 100, 2*R_SUN)
        
        t_init = fuori.temps[0]
        print(t_init)
        if mod:
            fuori.temps = convertToPwrLaw(fuori.temps, fuori.r_a, 2*R_SUN)
            # 0.9 is corrective factor so Tmax lines up!
            fuori.temps = fuori.temps*(t_init/fuori.temps[0]) - 200
            fuori.temps = np.round(fuori.temps/100)*100
        fuori.createAnnuliFromDirec(grav, direc, 'stellar atmospheres')
        fuori.createModelSpectrum('stellar atmospheres', False)
        self.fuori.append(fuori)
    def createFUOriBB(self, r_star, r_outer, m_star, m_dot, inc, dist, grav, direc, a_v, atm_table, pl_index, mod):
        fuori = FUOri(r_star, r_star, r_outer, m_star, m_dot, np.radians(inc), a_v, atm_table, pl_index)
        fuori.setAnnuliValues(10000, 1000, 100, 2*R_SUN)
        t_init = fuori.temps[0]
        print(t_init)
        if mod:
            fuori.temps = convertToPwrLaw(fuori.temps, fuori.r_a, 2*R_SUN)
            # 0.9 is corrective factor so Tmax lines up!
            fuori.temps = fuori.temps*(t_init/fuori.temps[0]) - 200
            fuori.temps = np.round(fuori.temps/100)*100
        fuori.createAnnuliFromDirec(grav, direc, 'blackbodies')
        fuori.createModelSpectrum('stellar atmospheres', False)
        self.fuori.append(fuori)
    def createStar(self, r_star, t_star, dist, grav, a_v, atm_table):
        wav, lum = atm_table[(t_star, grav)][0], atm_table[(t_star, grav)][1]
        lum = pyasl.unred(wav, lum, -a_v/3.1)
        flux = lum*((r_star/dist)**2)
        self.star.append((wav, flux))
```
